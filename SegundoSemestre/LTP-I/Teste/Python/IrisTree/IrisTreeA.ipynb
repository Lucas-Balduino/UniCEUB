```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import graphviz
from yellowbrick.classifier import ConfusionMatrix
```

```python
df = pd.read_csv('iris.csv')
df
```

```python
#df.shape # mostra quantidade de linhas e colunas
#df.head(5) # mostra as x primeiras linhas
#df.tail(5) # mostra as x ultimas linhas
#df['class'] # mostra a coluna
#df['class'].value_counts() # mostra a quantidade de ocorrencias da classe
#df['class'].unique() # mostra ocorrencias unicas
#df['class'].nunique() # mostra quantidade de ocorrencias unicas
#df['sepal width']
df.iloc[3:5] #mostra apenas a linha x
```

```python
previsores = df.iloc[:,0:4].values
#print(previsores)
classe = df.iloc[:,4].values
print(classe)
```

```python
X_treinamento, X_teste, Y_treinamento, Y_teste = train_test_split(previsores,
                                                                  classe,
                                                                  test_size = 0.3,
                                                                  random_state =123)
Y_teste
#len(Y_treinamento)
```

```python
arvore = DecisionTreeClassifier()
arvore.fit(X_treinamento, Y_treinamento)
```

```python
export_graphviz(arvore, out_file='tree.dot')
with open('tree.dot') as file:
    dot_graph = file.read()
graphviz.Source(dot_graph)
```

```python
previsoes = arvore.predict(X_teste)
previsoes
```

```python
taxa_acerto = accuracy_score(Y_teste, previsoes)
taxa_acerto
```

```python
cm = ConfusionMatrix(DecisionTreeClassifier())
cm.fit(X_treinamento,Y_treinamento)
cm.score(X_teste, Y_teste)
cm.poof()
```
