```python
import matplotlib.pyplot as plt
import nltk
nltk.download('stopwords')
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import stopwords
from matplotlib.colors import ListedColormap
from wordcloud import WordCloud
import string
```

```python
corpus = PlaintextCorpusReader('C:/Users/lucas.balduino/Documents/CodigosLTP1-Python/WordCloud/WordCloud2/ArquivosTxt','.*',encoding = "UTF-8")
```

```python
arquivos = corpus.fileids()
#arquivos[14]
arquivos[0:10]
```

```python
texto = corpus.raw('53.txt')
texto
```

```python
todo_texto = corpus.raw()
todo_texto
```

```python
palavras = corpus.words()
print(palavras[170])
print(len(palavras))
print(len(arquivos))
```

```python
stops = stopwords.words('english')
stops
```

```python
palavras_semstop = [p for p in palavras if p not in stops]
print(len(palavras_semstop))
palavras_sem_pontuacao = [p for p in palavras_semstop if p not in string.punctuation]
print(len(palavras_sem_pontuacao))
```

```python
frequencia = nltk.FreqDist(palavras_sem_pontuacao)
mais_comuns = frequencia.most_common(100)
mais_comuns
```

```python
mapa_cores = ListedColormap(['orange', 'green', 'red', 'magenta'])

cloud = WordCloud(background_color= 'white',
                  colormap= mapa_cores,
                  stopwords = stops,
                  max_words = 100,)
cloud.generate(todo_texto)
plt.figure( figsize=(15,15) )
plt.imshow(cloud)
```
